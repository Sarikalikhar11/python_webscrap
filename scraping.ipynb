```python
#importing important libraries for web scraping

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
```

```python
#reading the url

httpObject = urlopen("https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1")
webdata = httpObject.read()
print(webdata)
```

```python
soupdata = soup(webdata)
type(soupdata)
```

```python
containers = soupdata.findAll('div',{'class':'s-card-container'})
print(type(containers),len(containers))
```

```python
containers[1]    #data of first container
```

```python
containers[0]
```

```python

f = open('amazonweb_scraps.csv','wb')
f.write('productname,stars,reviews,currentprice,MRP,imageurl\n'.encode())

for container in containers:
    #finding product name
    product = container.find('span',{'class':'a-size-medium a-color-base a-text-normal'})
    productname = (container.text.split('4')[0].strip()).replace('/','')

    #finding rating
    star = container.find('i',{'class':'a-icon'})
    try:
        stars = star.text.replace('out of 5','')
    except:
        stars = 0

    #finding reviews
    review = container.find('span',{'class':'a-size-base'})
    try:
        reviews = review.text
    except:
        reviews = 0

    #finding price
    price = container.find('span',{'class':'a-price-whole'})
    try:
        currentprice = price.text.replace(',','')
    except:
        currentprice = 0
    
    #finding discounted price
    mrp = container.find('span',{'class':'a-offscreen'})
    try:
        MRP = mrp.text.replace(',','')
    except:
        MRP = 0

    #finding image url
    image = container.img
    imageurl = image.get('src')

    print(productname, stars, reviews, currentprice, MRP, imageurl)
    f.write(f"{productname},{stars},{reviews},{currentprice},{MRP},{imageurl}\n".encode())
    print('\n')
f.close()
```
